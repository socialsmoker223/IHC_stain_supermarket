```python
import requests
import io
import pandas as pd
import json
```

```python
def run(csv):
    
    url = 'http://127.0.0.1:8000/train_rules'
    files = {"file": open(csv, 'rb')}
    r = requests.post(url, files=files)
    data = r.text
    df = pd.DataFrame(json.loads(data)['df'])
    return df
```

```python
df = run('./data/ihc_record.csv')
df
```
